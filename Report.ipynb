# Face Mask with Social Distance Detector

This is a documentation on a detector created to help alleviate enforcing the current safety regulations. As we all know, the current world is experiencing a pandemic known as the Covid-19. So far, almost all countries have in some way or another enforced rules and regulations to help spread the virus. Out of all the rules and regulations, the most common ones are mandatory face mask and the practice of social distancing of a minimum of 6 feet apart. Although this pandemic has been ongoing for almost a year, people are still either forgetting or ignoring the stated guidelines. Therefore, I plan to create a detector that can be implemented into a simple device placed strategically at the entrances of buildings to help remind the people to wear their mask and maintain social distance.

# ML Models

- Convolutional Neural Network (ConvNet/CNN)
  - Consist of neural networks connected by neurons that works in its own receptive field to process data in a grid-like topology
- InceptionV3
  - A CNN network architecture that uses label smoothing, factorized 7x7 convolutions, and auxiliary classifier to transfer label information along the network
- MobileNet
  - Designed for mobile and embedded vision applications by using depthwise separable convolutions to build deep neural network while having low-latency
- DenseNet
  - As the name suggest it uses dense blocks, where dense connections between layers are made to connect all layers directly with each other
- VGG19
  - Consists of 19 layers: 16 concolution layers, 3 fully connected layers, 5 maxpooling layers, and 1 softmax layer

# Improvements / Suggestions

The face detector model above is not very sensitive due to the faces being mostly covered. To improve on the model, I should make full use of the haarcascade data by matching not only the face but the eyes and profile. This will help improve the face recognition even when only the eyes are detected as shown in the picture above.
